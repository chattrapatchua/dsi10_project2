### Feature Engineering:


#### *- Identify columns that can be deleted:*

```python
# Drop ID, PID, Garage Cars
# Id column has not correlation with the value
# PID is related to location, which is given by other variables already
# Garage Cars is represented by Garage Area already
 
train.drop(labels=['Id', 'PID', 'Garage Cars'], axis=1, inplace=True)
test.drop(labels=['PID', 'Garage Cars'], axis=1, inplace=True)
```

```python
# Pool Area and Pool QC can be dropped
# Only few houses with pools, and from EDA done the SalePrice has negligible correlation with whether there is a pool or not

train.drop(labels=['Pool Area', 'Pool QC'], axis=1, inplace=True)
test.drop(labels=['Pool Area', 'Pool QC'], axis=1, inplace=True)
```

```python
# Drop Garage Qual, Exter Qual, Bsmt Qual, Kitchen Qual
# There is a variable for Overall Qual already

train.drop(labels=['Garage Qual', 'Exter Qual', 'Bsmt Qual', 'Kitchen Qual'], axis=1, inplace=True)
test.drop(labels=['Garage Qual', 'Exter Qual', 'Bsmt Qual', 'Kitchen Qual'], axis=1, inplace=True)
```

#### *- Identify rows that can be deleted:*

```python
# Drop all rows where MS Zoning = A (agr) & I (all) 
# These classifications corresponds to agricultural & industrial zoning respectively
# Should only consider houses

train.drop(train[(train['MS Zoning'] == 'A (agr)')| (train['MS Zoning'] == 'I (all)')].index, inplace=True)
#test.drop(test[(test['MS Zoning'] == 'A (agr)')| (test['MS Zoning'] == 'I (all)')].index, inplace=True)
```

#### - *Combining features:*

```python
# Combine all variables related to size of house to get Total_Living_SF
# Total Bsmt SF, Gr Liv Area, Wood Deck SF, Open Porch SF, 3Ssn Porch
# Drop the variables used for Total_Living_SF due to high correlation

train['Total_Living_SF'] = (train['Total Bsmt SF'] + train['Gr Liv Area'] + train['Wood Deck SF'] 
                            + train['Open Porch SF'] + train['3Ssn Porch'] + train['Enclosed Porch'] + train['Low Qual Fin SF']
                            + train['Screen Porch'])
        
train.drop(labels=['BsmtFin SF 1', 'BsmtFin SF 2','Bsmt Unf SF', 'Total Bsmt SF', 
                   'Gr Liv Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch',
                   'Screen Porch', '3Ssn Porch', '1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF'], 
           axis=1, inplace=True)

test['Total_Living_SF'] = (test['Total Bsmt SF'] + test['Gr Liv Area'] + test['Wood Deck SF'] 
                            + test['Open Porch SF'] + test['3Ssn Porch'] + test['Enclosed Porch'] + test['Low Qual Fin SF']
                            + test['Screen Porch'])
        
test.drop(labels=['BsmtFin SF 1', 'BsmtFin SF 2','Bsmt Unf SF', 'Total Bsmt SF', 
                   'Gr Liv Area', 'Wood Deck SF', 'Open Porch SF', 'Enclosed Porch',
                   'Screen Porch', '3Ssn Porch', '1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF'], 
           axis=1, inplace=True)
```

```python
# Drop variables that have multicollinearity with Total_Living_SF
lst = ['Bsmt Full Bath', 'Bsmt Half Bath', 'TotRms AbvGrd', 'Half Bath', 'Full Bath',
      'Kitchen AbvGr', 'Bedroom AbvGr']
train.drop(labels=lst, axis=1, inplace=True)
test.drop(labels=lst, axis=1, inplace=True)
```

#### *- Dealing with Null Values:*

```python
# sort by columns with most null values
train.isnull().sum().sort_values(ascending=False).head(30)
```

*- Identify variables where NaN values cannot be removed:*

```python
# List of all the categorical columns with 'NA'/'None' etc. as a category
lst_cols = ['Misc Feature',
            'Alley',
            'Fence',
            'Fireplace Qu',
            'Garage Finish',
            'Garage Cond',
            'Garage Yr Blt',
            'Garage Type',
            'Bsmt Exposure',
            'BsmtFin Type 2',
            'BsmtFin Type 1',
            'Bsmt Cond',
            'Mas Vnr Type',
            'Lot Frontage']
```

```python
# Define fn which converts NaN values to 'None' string
def nan_converter(df, lst_cols): # Define fn which converts NaN values to 'None' string
    for col in lst_cols:
        if df[col].dtypes == 'O': # if dtype is string, then replace NaN with 'None'
            df[col].fillna('None', inplace=True)
        else:
            df[col].fillna(0, inplace=True) # otherwise, replace with int(0)
```

```python
# Call the nan_converter fn
nan_converter(train, lst_cols)
```

```python
# Find any other columns with null values
train.isnull().sum().sort_values(ascending=False).head(30)
```

```python
# For remaining NaN values, drop the whole row
# They are not categorical variables that have 'None' categories
# The number of dropped rows is relatively small, therefore not much of an impact
train = train.dropna()
```

### Explore Categorical Variables

#### - Groupby
- Rather than use one-hot encoding and creating numerous dummy variables (clutters up the dataframe)
- Groupby the categories and calculate the mean SalePrice for each category in the categorical variables
    - In the modelling section, all the variables will be scaled using the z-score
- In effect, this achieves the same purpose as using one-hot encoding

```python
# Location is given by the Neighborhood variable.
# Location should have high correlation with SalePrice.
# Groupby on Neighborhood and find mean price of each neighborhood. 
# Create new column with the mean price corresponding to the neighborhood 'Neighborhood_MeanPrice'

train.groupby('Neighborhood').mean()['SalePrice'].sort_values().plot(kind='barh', figsize=(7,7))
train['Neighborhood_MeanPrice'] = train['Neighborhood'].map(train.groupby('Neighborhood').mean()['SalePrice'])
train.head()
```

- Prices vary quite a lot by Neighborhood. StoneBr is the most expensive neighborhood, whilst MeadowV is cheapest.
- Shows that the Neighborhood (i.e. location) has strong influence on the SalePrice

```python
# Assign the mean price for each neighborhood to a dict
# This is so that these values can be mapped to the test dataset
# Subsequently, create a Neighborhood_MeanPrice column in the test dataset with the meanprices associated w/ each neighborhood
neighborhood_meanprice_dict = dict(train.groupby('Neighborhood').mean()['SalePrice'])
test['Neighborhood_MeanPrice'] = test['Neighborhood'].map(lambda x: neighborhood_meanprice_dict[x])
```

```python
# define fn to groupby all categorical variables with dtype == str, and show mean SalePrice, median, and count

def groupby_categorical_str(df):
    for col in df:
        if (df[col].dtypes == 'O') | (col == 'MS SubClass'):
            df.groupby(col)['SalePrice'].agg(['mean', 'median']).sort_values(by='mean').plot(kind='bar')
```

```python
plt.rcParams.update({'figure.max_open_warning': 0}) # prevent the max_open warning from appearing so that window does not get clogged up
groupby_categorical_str(train)
```

```python
def new_groupby_col(df):
    for col in df:
        if (df[col].dtypes == 'O') | (col == 'MS SubClass'):
            df[col+'_MeanPrice'] = df[col].map(df.groupby(col)['SalePrice'].mean())
```

```python
new_groupby_col(train)
```

## EDA of choice predictor variables

```python
# Scatterplot of each 
# Define fn to subplot the scatterplots

def subplot_scatterplt(dataframe, list_of_plot): 
    nrows = int(np.ceil(len(list_of_plot)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(10,10)) # You'll want to specify your figsize
    plt.subplots_adjust(hspace=0.4, wspace=0.5) # Assign certain distance btw subplot rows 
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, plot in enumerate(list_of_plot): # Gives us an index value to get into all our lists
        ax[i].scatter(dataframe[list_of_plot[i][0]], dataframe[list_of_plot[i][1]]) 
        # Set titles, labels, etc here for each subplot
        ax[i].set_title(list_of_plot[i][0] + ' vs. ' + list_of_plot[i][1], size=15)
        ax[i].set_xlabel(list_of_plot[i][0], size=15)
        ax[i].tick_params(axis='both', which='major', labelsize=15)
        ax[i].set_ylabel(list_of_plot[i][1], size=15)        
        # calc and plot the trendline
        z = np.polyfit(dataframe[list_of_plot[i][0]], dataframe[list_of_plot[i][1]], 1)
        p = np.poly1d(z)
        ax[i].plot(dataframe[list_of_plot[i][0]],p(dataframe[list_of_plot[i][0]]), "r--" )
```

```python
list_of_plot = [['Overall Qual', 'SalePrice'], ['Total_Living_SF', 'SalePrice'], ['Garage Area', 'SalePrice'],
               ['Neighborhood_MeanPrice', 'SalePrice'], ['Year Built', 'SalePrice']]
subplot_scatterplt(train, list_of_plot)
```

# Model Construction

```python
features = ['Overall Qual', 'Total_Living_SF', 'Garage Area', 'Neighborhood_MeanPrice', 'Year Built', 'Overall Cond']
```

```python
test[features].tail()
```

## Scaling

```python
X_train = train[features]
X_test = test[features]
y = train['SalePrice']
ss = StandardScaler()
ss.fit(X_train)
X_train_tf = ss.transform(X_train)
X_train = pd.DataFrame(X_train_tf, columns=X_train.columns)
X_test_tf = ss.transform(X_test)
X_test = pd.DataFrame(X_test_tf, columns=X_test.columns)
```

```python
X_train.head()
```

```python
X_test.tail()
```

```python
def lr_model_train(df=train, features=[]): # takes the dataframe, and the features list
    X = df.loc[:, features] # Subsets the dataframe to only include the features in the list, and assign to X
    y = train['SalePrice']  # the target variable y is the SalePrice
    lr = LinearRegression()
    
    # Model using linear regression
    model = lr.fit(X, y)
    predictions = model.predict(X)
    print("Model RSquared score:", model.score(X, y))
    print("RMSE:", np.sqrt(metrics.mean_squared_error(y, predictions)))
    
    # Use cross validation to assess model
    scores = cross_val_score(model, X, y, cv=10, scoring='r2')
    print("Cross-validation RSquared scores:", scores)
    print("Cross-validation mean RSquared score:", scores.mean())
    
    # Plot the predictions vs. the actual values for the target variable
    plt.scatter(predictions, y)
    ax = plt.gca()
    z = np.polyfit(predictions, train['SalePrice'], 1)
    p = np.poly1d(z)
    ax.plot(predictions,p(predictions), "r--" )
    ax.set_xlabel('Predicted SalePrice')
    ax.set_ylabel('Actual SalePrice')
    
    return model
```

### Model 1: Baseline Linear Regression model with single predictor variable (SLR)

```python
# Total_Living_SF was chosen as the baseline model predictor, as it is the most intuitive variable influencing SalePrice
lr_model_train(features=['Total_Living_SF'])
```

- The baseline SLR model with 'Total_Living_SF' as the predictor
- The R^2 score is 0.676
- Cross-validation R^2 score (mean) for K-fold = 10 is 0.614, showing a relatively stable model

### Model 2: Linear Regression w/ Multiple predictor variables (MLR) w/ 4 predictors

```python
# Picked 4 predictor variables according to strength of correlation to SalePrice (based on heatmap)
lr_model_train(features=['Overall Qual', 'Total_Living_SF', 'Garage Area', 'Neighborhood_MeanPrice'])
```

```python
# Calculate change in RMSE from baseline model to Model 2
(45011 - 31439)/45011
```

- Increasing the number of predictor variables increased the R^2 score quite significantly
- It also reduced the root mean square error (RMSE) by around 30% from baseline model
- Cross-validation mean score is 0.84, suggesting a model that has not been overfit

## Ridge Regularization

- Ridge regularization was chosen in an attempt to see if regularization would improve the model's performance.
- The predictor variables used were the same as from Model 6
- Lasso regularization was not chosen as feature selection has been manually done already, and thus forcing some variable coefficients to 0 may result in higher RMSE when it comes to testing.

```python
def ridge_model_train(df=X_poly, features=list(X_poly.columns)): # takes the dataframe, and the features list
    X = df.loc[:, features] # Subsets the dataframe to only include the features in the list, and assign to X
    y = train['SalePrice']  # the target variable y is the SalePrice

    # Model using Ridge
    alphas = np.logspace(0, 5, 200)
    model = RidgeCV(alphas=alphas, store_cv_values=False)
    model = model.fit(X, y)
    optimal_alpha = model.alpha_
    model_opt = Ridge(alpha=optimal_alpha)
    model_opt.fit(X, y)
    predictions = model_opt.predict(X)
    print("Model RSquared score:", model_opt.score(X, y))
    print("RMSE:", np.sqrt(metrics.mean_squared_error(y, predictions)))
    # Use cross validation to assess model
    scores = cross_val_score(model_opt, X, y, cv=10, scoring='r2')
    print("Cross-validation RSquared scores:", scores)
    print("Cross-validation mean RSquared scores:", scores.mean())
    # Plot the predictions vs. the actual values for the target variable
    plt.scatter(predictions, y)
    ax = plt.gca()
    z = np.polyfit(predictions, train['SalePrice'], 1)
    p = np.poly1d(z)
    ax.plot(predictions,p(predictions), "r--" )
    ax.set_xlabel('Predicted SalePrice')
    ax.set_ylabel('Actual SalePrice')
    
    return model
```
