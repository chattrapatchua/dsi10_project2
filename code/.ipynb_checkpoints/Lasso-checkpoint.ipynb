```python
# Import libraries
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn import metrics
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV, ElasticNet, ElasticNetCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 500)
train = pd.read_csv('../datasets/train.csv')
train.head()
```

```python
# Get rid of whitespace before and after the column names
train.columns = train.columns.str.strip()
```

```python
# Pool Area and Pool QC can be dropped
# Only few houses with pools, and from EDA done the SalePrice has negligible correlation with whether there is a pool or not

train.drop(labels=['Pool Area', 'Pool QC'], axis=1, inplace=True)
```

```python
# Drop Garage Qual, Exter Qual, Bsmt Qual, Kitchen Qual
# There is a variable for Overall Qual already

train.drop(labels=['Garage Qual', 'Exter Qual', 'Bsmt Qual', 'Kitchen Qual'], axis=1, inplace=True)
```

```python
# List of all the categorical columns with 'NA'/'None' etc. as a category
lst_cols = ['Misc Feature',
            'Alley',
            'Fence',
            'Fireplace Qu',
            'Garage Finish',
            'Garage Cond',
            'Garage Yr Blt',
            'Garage Type',
            'Bsmt Exposure',
            'BsmtFin Type 2',
            'BsmtFin Type 1',
            'Bsmt Cond',
            'Mas Vnr Type',
            'Lot Frontage']
```

```python
# Define fn which converts NaN values to 'None' string
def nan_converter(df, lst_cols): # Define fn which converts NaN values to 'None' string
    for col in lst_cols:
        if df[col].dtypes == 'O': # if dtype is string, then replace NaN with 'None'
            df[col].fillna('None', inplace=True)
        else:
            df[col].fillna(0, inplace=True) # otherwise, replace with int(0)
```

```python
# Call the nan_converter fn
nan_converter(train, lst_cols)
```

```python
train.isnull().sum()
```

```python
# For remaining NaN values, drop the whole row
# They are not categorical variables that have 'None' categories
# The number of dropped rows is relatively small, therefore not much of an impact
train = train.dropna()
```

```python
def dummies_creator(df):
    for col in df:
        if df[col].dtypes == 'O':
            df = pd.get_dummies(data=df, columns=[col], drop_first=True)
    
    return df
```

```python
dummies_creator(train)
```

```python
train.info()
```

```python
target = 'SalePrice'
pred = [x for x in train.columns if (x != target) and (train[x].dtypes != 'O')]
```

```python
train[pred] = (train[pred] - train[pred].mean())/ train[pred].std()
```

```python
l_alphas = np.arange(0.001, 0.15, 0.0025)
lasso_model = LassoCV(alphas=l_alphas, cv=5)
lasso_model = lasso_model.fit(train[pred], train[target])
```

```python
lasso_optimal_alpha = lasso_model.alpha_
lasso_optimal_alpha
```

```python
lasso_model = Lasso(alpha=lasso_optimal_alpha)
lasso_model.fit(train[pred], train[target])
predictions = lasso_model.predict(train[pred])
print("Model RSquared score:", lasso_model.score(train[pred], train[target]))
print("RMSE:", np.sqrt(metrics.mean_squared_error(train[target], predictions)))
```
